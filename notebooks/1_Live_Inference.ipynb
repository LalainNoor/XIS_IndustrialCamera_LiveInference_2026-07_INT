# Industrial Camera Setup & Live Inference

## Objective

Develop a simulated industrial camera live inference pipeline using OpenCV and ONNX Runtime.

Since GenICam hardware was unavailable during development, a prerecorded video was used to simulate the camera stream. The software architecture was designed to allow replacement of the video source with a GenICam camera using the Harvester library in future deployments.

---

### Pipeline

Input Video
→ Camera Module
→ Frame Preprocessing
→ ONNX Runtime
→ Scene Classification
→ Live Overlay
→ Performance Metrics

## Project Components

- Camera Module (`camera.py`)
- Preprocessing (`preprocess.py`)
- ONNX Runtime Inference (`inference.py`)
- Display Overlay (`display.py`)
- Performance Metrics (`metrics.py`)
- Main Application (`main.py`)
